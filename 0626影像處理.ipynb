```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
text = open('mayday1.txt', encoding='utf-8').read()
wc = WordCloud(background_color='white', max_words = 200)
wc.generate(text)
plt.imshow(wc)
plt.axis('off')
plt.show()
wc.to_file('wordcloud1.jpg')
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
text = open('mayday1.txt', encoding='utf-8').read()
stopwords = {}.fromkeys(['yeah','oh','','la'])
print(stopwords)
wc = WordCloud(background_color='white', max_words = 300, font_path='NotoSerifCJKtc-Black.otf', stopwords = stopwords)
wc.generate(text)
plt.imshow(wc)
plt.axis('off')
plt.show()
wc.to_file('wordcloud3.jpg')
```

```python
import jieba
import collections
print('====stopWords.txttxt====')
list1 = []
stopWords = []  # 
remainderWords = []  # 
with open('stopWords.txt','r', encoding='utf-8') as file:
    for data in file:
        data = data.strip()  # 
        stopWords.append(data)
with open('data.txt','r', encoding='utf-8') as file:
    text = file.read()
    words = jieba.lcut(text)
print(words)
print('')
word_freq = collections.Counter(words)
print(word_freq)
```

```python
word_freq = {'Python':100,'Data':80,'Visualization':60,'Machine':50,'Learning':40}
wc = WordCloud(background_color='white', font_path='NotoSerifCJKtc-Black.otf')
wc.generate_from_frequencies(word_freq)
plt.imshow(wc)
plt.axis('off')
plt.show()
wc.to_file('wordcloud5.jpg')
```
